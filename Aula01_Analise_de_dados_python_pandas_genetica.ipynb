# Desafio 07 

Renomear as colunas tirando o hífen

Passando os nomes das colunas para string 

```python
dados.columns = dados.columns.astype('str')
```

removendo todos os espeços da direita e da esquerda 

```python
dados.columns = dados.columns.str.strip()
```

```python
dados.columns = dados.columns.str.lower()
```

substituindo os - por _

```python
dados.columns = map(lambda x : x.replace("-", "_").replace(" ", "_"), dados.columns)
dados
```
